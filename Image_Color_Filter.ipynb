```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
```

```python
def image_color_filter():
  imagesource1 = input('What image do you want to be filtered?')
  valid = 0
  while valid == 0:
    try:
      image1 = skimage.io.imread(str(imagesource1))
      valid = 1
    except SyntaxError:
      input("Please put in an url with http:// and all that jazz.")

  #Filtering images to each condition

  image1_r, image1_g, image1_b = np.copy(image1),np.copy(image1),np.copy(image1)
  image1_rb, image1_bg, image1_rg = np.copy(image1),np.copy(image1),np.copy(image1)

  image1_r[:,:,(1,2)], image1_g[:,:,(0,2)], image1_b[:,:,(0,1)] = 0, 0, 0
  image1_rb[:,:,(1)], image1_bg[:,:,(0)], image1_rg[:,:,(1)] = 0, 0, 0

  # Plotting the images in each color in one sheet (with titles).
  # I could have concactanated the arrays, but that would not have allow me to add labels on each image.

  f, ax = plt.subplots(3,3,tight_layout=True)
  axs = plt.gca()
  axs.get_yaxis().set_visible(False)
  ax[0,1].imshow(image1)
  ax[0,1].set_title("Original")

  ax[1,0].imshow(image1_r)
  ax[1,0].set_title("Red")

  ax[1,1].imshow(image1_g)
  ax[1,1].set_title("Green")

  ax[1,2].imshow(image1_b)
  ax[1,2].set_title("Blue")

  ax[2,0].imshow(image1_rg)
  ax[2,0].set_title("Red and Green")

  ax[2,1].imshow(image1_bg)
  ax[2,1].set_title("Blue and Green")

  ax[2,2].imshow(image1_rb)
  ax[2,2].set_title("Red and Blue")

  for i in range(0,3):
    b = i
    ax[b,0].axis("Off")
    ax[b,1].axis("Off")
    ax[b,2].axis("Off")
```

```python
image_color_filter()
```
